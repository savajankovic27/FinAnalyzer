```python
# 01_exploratory_analysis.ipynb

import sys
import os

module_path = os.path.abspath(os.path.join('..', 'Scripts'))
if module_path not in sys.path:
    sys.path.append(module_path)

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from data_loader import load_stock_data
from analysis_utils import compute_summary_statistics, add_returns
from technical_indicators import moving_average, exponential_moving_average, relative_strength_index
from plotting_utils import plot_prices

#Data Load... testing AAPL
symbol = "AAPL"
df = load_stock_data(symbol, data_dir="../Data")  


display(df.head())
display(df.tail())
print(df.info())
print("Summary Statistics:")
display(compute_summary_statistics(df))

#Dataframe indicators
# Had issues implementing the previous dataframes so i removed and replaced, seems to work. 
df = add_returns(df)
df = moving_average(df, window=20)
df = exponential_moving_average(df, span=20)
df = relative_strength_index(df, window=14)

display(df[['Adj Close', 'Daily_Return', 'SMA_20', 'EMA_20', 'RSI']].head(25))

## Visualization
plot_prices(df, symbol, show_sma=[20], show_ema=[20])

# RSI Plot
plt.figure(figsize=(12,4))
plt.plot(df['RSI'], label='RSI')
plt.axhline(70, color='red', linestyle='--', label='Overbought Threshold')
plt.axhline(30, color='green', linestyle='--', label='Oversold Threshold')
plt.title(f"{symbol} RSI")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.grid(True)
plt.show()


```
